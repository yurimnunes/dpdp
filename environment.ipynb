```python
# pickle read gdown tsp100_validation_seed4321.pkl
import pickle
import numpy as np
import matplotlib.pyplot as plt
#data = pickle.load(open('tsp100_test_seed1234.pkl', 'rb'))
# read tsp10_test_concorde.txt txt file
with open('tsp10_test_concorde.txt') as f:
    data = f.read()

data = data.split('\n')
```

```python
for i in range(len(data)):
    interesse = data[i].split()
    interesse = interesse[:20]
    data[i] = [float(x) for x in interesse]
    
```

```python
# for each entry in data, transform the 1x20 into 2x10
for i in range(len(data)):
    if data[i] != []:
        data[i] = np.array(data[i]).reshape(10,2)
```

```python
data
```

```python
for i in range(len(dados_solucao)):
    interesse = dados_solucao[i].split()
    interesse = interesse[21:]
    dados_solucao[i] = [int(x) for x in interesse]
    
```

```python
dados_solucao
```

```python
# write dados to txt file
with open('tsp10_data.txt', 'w') as f:
    for item in data:
        f.write("%s\n" % item)

# write dados_solucao to txt file
with open('tsp10_solucao.txt', 'w') as f:
    for item in dados_solucao:
        f.write("%s\n" % item)

# save solucao to pkg
solucao = dados_solucao
pickle.dump(solucao, open('tsp10_solucao.pkl', 'wb'))

# save train to pkl
train = data
pickle.dump(train, open('tsp10_data.pkl', 'wb'))
```

```python
for i in range(len(dados_solucao)):
    dados_solucao[i] = dados_solucao[i][:-1]
    # subtract 1 from each element in the array
    dados_solucao[i] = [x-1 for x in dados_solucao[i]]
```

```python
# sava data to np.save('array.npy', array)
data = data[:-1]
dados_solucao = dados_solucao[:-1]


# Calculate the split indices
num_samples = len(data)
train_end = int(0.7 * num_samples)
val_end = int(0.9 * num_samples)

# Split the data
train_data = data[:train_end]
solucao_train = dados_solucao[:train_end]
val_data = data[train_end:val_end]
solucao_val = dados_solucao[train_end:val_end]
test_data = data[val_end:]
solucao_test = dados_solucao[val_end:]

# Save each split
np.save('tsp10_train.npy', train_data)
np.save('tsp10_val.npy', val_data)
np.save('tsp10_test.npy', test_data)
np.save('tsp10_solucao_train.npy', solucao_train)
np.save('tsp10_solucao_val.npy', solucao_val)
np.save('tsp10_solucao_test.npy', solucao_test)
```
